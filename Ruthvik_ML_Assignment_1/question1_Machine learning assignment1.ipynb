```python
## 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset from the CSV file
data = pd.read_csv('kmtest.csv', header=None)
data = np.array(data)

# Define the range of K values
k_values = [2, 3, 4, 5]

# Iterate through each K value and perform K-means clustering
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    clusters = kmeans.fit_predict(data)
    cluster_centers = kmeans.cluster_centers_

    # Plot the data points with different colors for each cluster
    plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='rainbow', alpha=0.5)
    
    # Plot the cluster centers
    plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', marker='x', s=100)
    
    plt.title(f'K-means Clustering (K={k})')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset from the CSV file
data = pd.read_csv('kmtest.csv', header=None)
data = np.array(data)

# Normalize the data using z-score normalization
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data)

# Define the range of K values
k_values = [2, 3, 4, 5]

# Iterate through each K value and perform K-means clustering on the normalized data
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    clusters = kmeans.fit_predict(normalized_data)
    cluster_centers = kmeans.cluster_centers_

    # Plot the data points with different colors for each cluster
    plt.scatter(normalized_data[:, 0], normalized_data[:, 1], c=clusters, cmap='rainbow', alpha=0.5)
    
    # Plot the cluster centers
    plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', marker='x', s=100)
    
    plt.title(f'K-means Clustering with Normalization (K={k})')
    plt.xlabel('Normalized X-axis')
    plt.ylabel('Normalized Y-axis')
    plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

iris_data=pd.read_csv("iris.csv")
# Extract the features (columns 1, 2, 3, and 4) from the Iris dataset
X = iris_data.iloc[:, :4].values

# Perform K-means clustering multiple times to get different results
best_result = None
best_inertia = np.inf
worst_result = None
worst_inertia = -np.inf

for _ in range(5):
    kmeans = KMeans(n_clusters=3, random_state=None)
    kmeans.fit(X)
    inertia = kmeans.inertia_
    
    if inertia < best_inertia:
        best_inertia = inertia
        best_result = kmeans
    if inertia > worst_inertia:
        worst_inertia = inertia
        worst_result = kmeans

# The best and worst clustering results are stored in best_result and worst_result.
```

```python
import matplotlib.pyplot as plt

# Extract cluster centers from best and worst results
best_cluster_centers = best_result.cluster_centers_
worst_cluster_centers = worst_result.cluster_centers_

# Predict cluster assignments for all data points
best_labels = best_result.labels_
worst_labels = worst_result.labels_

# Plot best clustering result
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 2], X[:, 3], c=best_labels, cmap='viridis')
plt.scatter(best_cluster_centers[:, 2], best_cluster_centers[:, 3], c='black', marker='x', s=200)
plt.title('Best Clustering Result')
plt.xlabel('Attribute 3')
plt.ylabel('Attribute 4')

# Plot worst clustering result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 2], X[:, 3], c=worst_labels, cmap='viridis')
plt.scatter(worst_cluster_centers[:, 2], worst_cluster_centers[:, 3], c='black', marker='x', s=200)
plt.title('Worst Clustering Result')
plt.xlabel('Attribute 3')
plt.ylabel('Attribute 4')

plt.tight_layout()
plt.show()
```
